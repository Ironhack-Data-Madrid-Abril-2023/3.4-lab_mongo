### 2. All the companies that have more than 5000 employees. Limit the search to 20 companies and sort them by **number of employees**.

```python
# Your Code
list(colec.find({'number_of_employees': {'$gt': 5000}}, {"name":True, "number_of_employees":True}).limit(20).sort("number_of_employees"))
#gt es greater than 
# gte es greter than and equal 
# lt less than
# imprimo solo los nombre y numeros de empleados
```

### 3. All the companies founded between 2000 and 2005, both years included. Retrieve only the `name` and `founded_year` fields.

```python
# Your Code
list(colec.find({"founded_year": {"$gte": 2000, "$lte": 2005}}, {"name": 1, "founded_year": 1, "_id": 0}).limit(20))
# lo limitito solo a 20 para simplificar la correccion 
```

### 4. All the companies that had a Valuation Amount of more than 100.000.000 and have been founded before 2010. Retrieve only the `name` and `ipo` fields.

```python
# Your Code
list(colec.find({"$and": [{"ipo.valuation_amount": {"$gt": 100000000}}, {"founded_year": {"$lt": 2010}}]}, {"name": 1, "ipo": 1, "_id": 0}).limit(5))
#elimito a 5 para hacer mas factil la correccion 
```

### 5. All the companies that have less than 1000 employees and have been founded before 2005. Order them by the number of employees and limit the search to 10 companies.

```python
# Your Code
list(colec.find({"$and": [{"number_of_employees": {"$lt": 1000}}, {"founded_year": {"$lt": 2005}}]}).sort("number_of_employees", 1).limit(10))
```

### 6. All the companies that don't include the `partners` field.

```python
# Your Code
list(colec.find({'partners': {'$nin': [0]}}, {'name': True, '_id': False}))
```

### 7. All the companies that have a null type of value on the `category_code` field.

```python
# Your Code
list(colec.find({"category_code": None}, {"name":True}).limit(5))
```

### 8. All the companies that have at least 100 employees but less than 1000. Retrieve only the `name` and `number of employees` fields.

```python
# Your Code
list(colec.find({"$and": [{"number_of_employees": {"$gte": 100}}, {"number_of_employees": {"$lt": 1000}}]}, {"name": 1, "number_of_employees": 1, "_id":0}))
```

### 11. All the companies founded on the second semester of the year. Limit your search to 1000 companies.

```python
# Your Code
list(colec.find({
    "founded_month": {"$gte": 7},
    "founded_month": {"$lte": 12}
}, {"name": 1, "_id":0}).limit(10))
```

### 12. All the companies founded before 2000 that have an acquisition amount of more than 10.000.00

```python
# Your Code
list(colec.find({
    "founded_year": {"$lt": 2000},
    "acquisition.price_amount": {"$gt": 10000000}
}, {"name": 1, "_id":0}).limit(10))
```

### 13. All the companies that have been acquired after 2010, order by the acquisition amount, and retrieve only their `name` and `acquisition` field.

```python
# Your Code
docs = colec.find({"acquisition.acquired_year": {"$gt": 2010}}, {"name": 1, "acquisition.price_amount": 1}).sort("acquisition.price_amount", -1).limit(10)

for doc in docs:
    print(doc["name"], doc["acquisition"]["price_amount"])
```

### 14. Order the companies by their `founded year`, retrieving only their `name` and `founded year`.

```python
# Your Code
docs = colec.find({}, {"name": 1, "founded_year": 1}).sort("founded_year", 1)

for doc in docs:
    print(doc["name"], doc["founded_year"])
```

### 15. All the companies that have been founded on the first seven days of the month, including the seventh. Sort them by their `acquisition price` in a descending order. Limit the search to 10 documents.

```python
# Your Code
list(colec.find({"founded_day": {"$lte": 7}}, {"_id": 0, "name": 1, "acquisition.price_amount": 1}).sort("acquisition.price_amount", -1).limit(10))
```

### 16. All the companies on the 'web' `category` that have more than 4000 employees. Sort them by the amount of employees in ascending order.

```python
# Your Code
list(colec.find({"category_code": "web", "number_of_employees": {"$gt": 4000}}, {"_id": 0, "name": 1, "number_of_employees": 1}).sort("number_of_employees", 1).limit(10))
```

### 17. All the companies whose acquisition amount is more than 10.000.000, and currency is 'EUR'.

```python
# Your Code
list(colec.find({"acquisition.price_amount": {"$gt": 10000000}, "acquisition.price_currency_code": "EUR"}, {"_id": 0, "name": 1, "acquisition": 1}).limit(3))
```

### 18. All the companies that have been acquired on the first trimester of the year. Limit the search to 10 companies, and retrieve only their `name` and `acquisition` fields.

```python
# Your Code
list(colec.find({"founded_month": {"$lte": 3}},{"_id": 0, "name": 1, "acquisition": 1}).limit(5))
```
